# Simple Sersic Galaxy

This example shows a very simple toy model of a pure disk ($n=1$) Sersic-profile galaxy at a moderate redshift obsered by METIS imaging and long-slit spectroscopy modes.

```python
from matplotlib import pyplot as plt
from astropy import units as u

from scopesim import Simulation
from scopesim_templates.extragalactic import galaxy3d
```

## Create source

This can take a few moments on slower machines...

```python
src = galaxy3d(
    "brown/NGC3773",
    z=7,
    vmax=800*u.km/u.s,
    sigma=20*u.km/u.s,
    theta=10*u.deg,
    r_eff=3*u.arcsec,
    n=1,
    ellip=.6,
    pixel_scale=.01,
    ngrid=20,
)
```

Based on the `ngrid` parameter, multiple source fields are created. Note that this number is _not_ identical to `ngrid`, but rather based on the distribution of $v$ and $\sigma$ within the model. A larger value of `ngrid` will however result in more smaller fields.

```python
len(src.fields)
```

### Plot source input

By summing up the individual source fields' data, we can reconstruct the full original image. In the following plot, red contour lines show the intensity profile of the image, whereas the white contour lines illustrate the borders of the individual fields. Rotational velocity and velocity dispersion are averaged within each field to produce an average spectrum based on the galaxy's overall scaled apectrum, effectively discretizing the galaxy.

```python
_, ax = plt.subplots(figsize=(10, 8))
img = sum(f.data.value for f in src.fields)
ax.imshow(img)
ax.contour(img, levels=10, colors="red", linewidths=.5)
for f in src.fields:
    ax.contour(f.data.value, levels=0, colors="white", linewidths=.5)
ax.invert_yaxis()
```

## Create and run simulations

### Imaging

First we will create a simulation for LM imaging mode. For the sake of simplicity, we will disable the atmosphere's influence and set the ambient temperature to "very cold", to avoid any thermal emissions in the results.

```python
sim_img = Simulation(
    "METIS",
    "img_lm",
    properties={
        "!OBS.exptime": 10000,  # seconds
        "!ATMO.temperature": -200,  # Celsius
        "!SIM.spectral.spectral_bin_width": 5.0e-5,
    },
    ignore_effects=["skycalc_atmosphere"],
)
```

Run simulation, this may take a moment on slower machines...

```python
result_img = sim_img(src)[1]
```

### Long-slit spectroscopy

Next we will look at LSS-L spectroscopy, with otherwise the same configuration.

```python
sim_lss = Simulation(
    "METIS",
    "lss_l",
    properties={
        "!OBS.exptime": 10000,  # seconds
        "!OBS.slit": "A-19_0",  # narrow slit
        "!ATMO.temperature": -200,  # Celsius
        "!SIM.spectral.spectral_bin_width": 5.0e-5,
    },
    ignore_effects=["skycalc_atmosphere"],
)
```

Run simulation, this may take a moment on slower machines...

```python
result_lss = sim_lss(src)[1]
```

### Slit image

For comparison and to visulalize where the slit ends up on-sky, let's run the imaging simulation again but with the slit in place.

```python
sim_img.optical_train["slit_wheel"].include = True
sim_img.optical_train["slit_wheel"].change_slit("A-19_0")
result_img_slit = sim_img(src)[1]
```

## Plot results

Now let's look at the two results side-by-side. For the spectroscopic observation, we'll look at the raw 2D spectrum as it comes out the detector.

```python
_, (ax0, ax1) = plt.subplots(1, 2, figsize=(12, 10))
ax0.imshow(result_img.data, origin="lower")
ax1.imshow(result_lss.data, origin="lower")
```

```python
from astropy.wcs import WCS
_, ax = plt.subplots(subplot_kw={"projection": WCS(result_img.header)}, figsize=(12, 10))
ax.imshow(result_img.data)
ax.contour(result_img_slit.data > 1e6, levels=0, colors="white", linewidths=.5)
ax.set_ylim(974, 1074)
ax.set_xlim(250, 1100)
```

```python
_, ax = plt.subplots(subplot_kw={"projection": WCS(result_lss.header)}, figsize=(10, 10))
ax.imshow(result_lss.data)
```

### Extract spectra

Finally, we will extract three vertical cuts through the 2D spectrum to show the differences along the long slit. Since the center is brighter than the outer regions, we'll roughly normalize the spectra to be comparable. Also, we'll limit the wavelength axis to a narrow window, to see the shifts better.

```python
_, ax = plt.subplots(figsize=(10, 4))
ax.plot(result_lss.data[:, 474] / 160000, label="left")
ax.plot(result_lss.data[:, 900] / 350000, label="left")
ax.plot(result_lss.data[:, 1024] / 500000, label="center")
ax.plot(result_lss.data[:, 1150] / 350000, label="right")
ax.plot(result_lss.data[:, 1574] / 160000, label="right")
ax.legend()
ax.set_xlim(350, 550)
# ax.set_ylim(0.7, 2.3)
```

```python
for s in src.spectra.values():
    plt.plot(src.spectra[15].waveset, s(src.spectra[15].waveset))
plt.xlim(52200, 52800)
```

```python
rect = sim_lss.optical_train["spectral_traces"].rectify_traces(sim_lss.last_readout[0], -4, 4)
```

```python
_, ax = plt.subplots(subplot_kw={"projection": WCS(rect[1].header)}, figsize=(10, 10))
ax.imshow(rect[1].data)
ax.set_xlim(1400, 1600)
ax.set_ylim(700, 762)
```

```python
import numpy as np
from scipy.optimize import curve_fit

swcs = WCS(rect[1].header).spectral
wave = swcs.pixel_to_world(np.arange(swcs.pixel_shape[0])) << u.um
```

```python
def gauss(x, a, b, c, d):
    return a * np.exp(-(x - b)**2/(2*c**2)) + d
```

```python
slc = slice(1510, 1550)
cenpx = 731
cuts = [-331, -231, -81, -31, 31, 81, 231, 331]
cuts = np.arange(-600, 600, 1)
xax = np.linspace(wave[slc].min(), wave[slc].max(), 200)
p0 = (350000, 4.005, .01, 1)

_, ax = plt.subplots(figsize=(10, 8))
ax.plot(wave[slc], rect[1].data[cenpx, slc], label="center")
popt_c, _ = curve_fit(gauss, wave[slc].value, rect[1].data[cenpx, slc], p0)
ax.plot(xax, gauss(xax.value, *popt_c), label="center fit", ls=":")
cenfit_c = popt_c[1] << u.um

cenfits = []
rs = []
for cut in cuts:
    xdata = wave[slc]
    ydata = rect[1].data[cenpx + cut, slc]
    # ax.plot(xdata, ydata, label=f"{cut} px")
    try:
        popt_cut, _ = curve_fit(gauss, xdata, ydata, p0)
    except RuntimeError:
        continue
    # ax.plot(xax, gauss(xax.value, *popt_cut), label=f"{cut} px fit", ls=":")
    cenfits.append(popt_cut[1] << u.um)
    rs.append(cenpx + cut)
cenfits <<= u.um
rs = np.array(rs)

ax.vlines(cenfits.value, 0, 3, color="k", linestyle=":")
ax.vlines(cenfit_c.value, 0, 3, color="k", linestyle="--")
ax.legend()
```

```python
c_0 = 299792458 << u.m/u.s
shifts = cenfits - cenfit_c << u.AA
vels = shifts / cenfit_c * c_0 << u.km/u.s
```

```python
vels
```

```python
plt.scatter(np.abs(rs - cenpx), np.abs(vels), s=10)
plt.xlabel("r [px]")
plt.ylabel("v [km/s]")
```
